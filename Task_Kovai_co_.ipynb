```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

```python
# Prepare data for time series analysis
df_ts = df.set_index('Date')['Total']

# Decompose time series
decomposition = seasonal_decompose(df_ts.resample('D').mean().fillna(method='ffill'),
                                  period=365, model='additive')

fig, axes = plt.subplots(4, 1, figsize=(15, 12))
decomposition.observed.plot(ax=axes[0], title='Observed')
decomposition.trend.plot(ax=axes[1], title='Trend')
decomposition.seasonal.plot(ax=axes[2], title='Seasonality')
decomposition.resid.plot(ax=axes[3], title='Residuals')
plt.tight_layout()
plt.show()
```

```python
# Calculate correlations between transport modes
correlation_matrix = df[total_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Between Transport Modes')
plt.show()

print("\nStrong Correlations (> 0.7):")
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.7:
            print(f"{correlation_matrix.columns[i]} - {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.3f}")
```

```python
# Analyze pre, during, and post COVID periods
covid_periods = {
    'Pre-COVID (2019)': ('2019-01-01', '2019-12-31'),
    'COVID (2020-2021)': ('2020-01-01', '2021-12-31'),
    'Post-COVID (2022-2024)': ('2022-01-01', '2024-09-30')
}

covid_analysis = {}
for period_name, (start_date, end_date) in covid_periods.items():
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    period_data = df.loc[mask]

    covid_analysis[period_name] = {
        'Avg Daily Total': period_data['Total'].mean(),
        'Avg Local Route': period_data['Local Route'].mean(),
        'Avg Light Rail': period_data['Light Rail'].mean(),
        'Avg Rapid Route': period_data['Rapid Route'].mean(),
        'Avg School': period_data['School'].mean(),
        'Avg Peak Service': period_data['Peak Service'].mean()
    }

covid_df = pd.DataFrame(covid_analysis).T
print("COVID Impact Analysis:")
print(covid_df)
```

```python
# Prepare data for forecasting
forecast_days = 7
last_date = df['Date'].max()
forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1),
                               periods=forecast_days, freq='D')

# Method 1: Simple Moving Average (Baseline)
def moving_average_forecast(data, window=30, forecast_days=7):
    """Simple moving average forecast"""
    return np.array([data[-window:].mean()] * forecast_days)

# Method 2: Exponential Smoothing (Holt-Winters)
def exponential_smoothing_forecast(data, forecast_days=7, seasonal_periods=365):
    """Exponential smoothing forecast"""
    model = ExponentialSmoothing(data, seasonal_periods=seasonal_periods,
                                trend='add', seasonal='add')
    fitted_model = model.fit()
    forecast = fitted_model.forecast(forecast_days)
    return forecast.values

# Method 3: Seasonal Naive (same day last year)
def seasonal_naive_forecast(data, dates, forecast_dates, seasonal_period=365):
    """Forecast based on same period last year"""
    forecasts = []
    for date in forecast_dates:
        # Find same date last year
        same_day_last_year = date - pd.Timedelta(days=seasonal_period)
        # Find closest date in historical data
        idx = (dates - same_day_last_year).abs().argmin()
        forecasts.append(data.iloc[idx])
    return np.array(forecasts)

# Generate forecasts for each transport mode
forecast_results = {}
models = ['Moving Average (30-day)', 'Exponential Smoothing', 'Seasonal Naive']

for mode in total_columns:
    data = df.set_index('Date')[mode]

    # Generate forecasts using different methods
    ma_forecast = moving_average_forecast(data.values, window=30, forecast_days=forecast_days)
    es_forecast = exponential_smoothing_forecast(data, forecast_days=forecast_days)
    sn_forecast = seasonal_naive_forecast(data, df['Date'], forecast_dates)

    # Store results
    forecast_results[mode] = {
        'MA': ma_forecast,
        'ES': es_forecast,
        'SN': sn_forecast,
        'Ensemble': (ma_forecast + es_forecast + sn_forecast) / 3  # Simple ensemble
    }

# Create forecast dataframe
forecast_df = pd.DataFrame({
    'Date': forecast_dates,
    **{f'{mode}_Forecast': forecast_results[mode]['Ensemble'] for mode in total_columns}
})

# Add total forecast
forecast_df['Total_Forecast'] = forecast_df[[f'{mode}_Forecast' for mode in total_columns]].sum(axis=1)

print("\n" + "="*50)
print("7-DAY FORECAST RESULTS")
print("="*50)
print(f"Forecast Period: {forecast_dates[0].strftime('%Y-%m-%d')} to {forecast_dates[-1].strftime('%Y-%m-%d')}")
print("\nDaily Forecasts:")
print(forecast_df[['Date', 'Total_Forecast'] + [f'{mode}_Forecast' for mode in total_columns]].round(0))

print("\n" + "="*50)
print("FORECAST SUMMARY STATISTICS")
print("="*50)

# Calculate forecast statistics
forecast_summary = pd.DataFrame({
    'Metric': ['Min', 'Max', 'Average', 'Std Dev', 'Week Total'],
    'Total': [
        forecast_df['Total_Forecast'].min(),
        forecast_df['Total_Forecast'].max(),
        forecast_df['Total_Forecast'].mean(),
        forecast_df['Total_Forecast'].std(),
        forecast_df['Total_Forecast'].sum()
    ]
})

# Add mode-specific statistics
for mode in total_columns:
    col_name = f'{mode}_Forecast'
    forecast_summary[mode] = [
        forecast_df[col_name].min(),
        forecast_df[col_name].max(),
        forecast_df[col_name].mean(),
        forecast_df[col_name].std(),
        forecast_df[col_name].sum()
    ]

print(forecast_summary.round(0))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset


# Choose one transport type, e.g., 'Local Route'
series = df['Local Route']

# Optional: fill missing values
series = series.fillna(method='ffill')

# Plot original series
series.plot(figsize=(12,5), title='Local Route Passengers')
plt.show()
```

```python
result = adfuller(series)
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

# If p-value > 0.05, series is non-stationary  consider differencing
```

```python

p, d, q = 1, 1, 1
P, D, Q, s = 1, 1, 1, 365

model = SARIMAX(series, order=(p,d,q), seasonal_order=(P,D,Q,s),
                enforce_stationarity=False, enforce_invertibility=False)
sarima_result = model.fit()
print(sarima_result.summary())
```

```python
forecast_days = 7
forecast = sarima_result.get_forecast(steps=forecast_days)
forecast_index = pd.date_range(start=series.index[-1] + pd.Timedelta(days=1), periods=forecast_days)
forecast_df = pd.DataFrame({'Forecast': forecast.predicted_mean.values}, index=forecast_index)

# Plot forecast
plt.figure(figsize=(12,5))
plt.plot(series, label='Historical')
plt.plot(forecast_df, label='Forecast', marker='o', color='red')
plt.title('SARIMA Forecast for Local Route')
plt.xlabel('Date')
plt.ylabel('Passengers')
plt.legend()
plt.show()

print(forecast_df)
```
